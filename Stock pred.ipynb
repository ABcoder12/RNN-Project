```python
import numpy as np
X_train = []
y_train = []

for i in range(5,922):
  X_train.append(scaled_train_set_df.iloc[i-5:i,0])
  y_train.append(scaled_train_set_df.iloc[i,0])

X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))
```

```python
X_train
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, GRU, SimpleRNN, Bidirectional
from keras.layers import Dropout
```

```python
model_1 = Sequential()
model_1.add(SimpleRNN(units = 50, return_sequences=True, input_shape = (X_train.shape[1],1), activation = 'tanh'))
model_1.add(Dropout(0.2))
model_1.add(SimpleRNN(units = 50, return_sequences=True, input_shape = (X_train.shape[1],1), activation = 'tanh'))
model_1.add(Dropout(0.2))
model_1.add(SimpleRNN(units = 50, return_sequences=True, input_shape = (X_train.shape[1],1), activation = 'tanh'))
model_1.add(Dropout(0.2))
model_1.add(SimpleRNN(units = 50, activation = 'tanh'))
model_1.add(Dropout(0.2))
model_1.add(Dense(units = 1))
model_1.summary()
```

```python
model_1.compile(optimizer = 'RMSProp', loss= 'mean_squared_error')
model_1.fit(X_train, y_train, epochs = 50, batch_size = 16)
```

```python
y_train_pred = model_1.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_train_pred)
```
